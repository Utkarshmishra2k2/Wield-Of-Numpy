Finding The Number Of Points Outside The N-Dimensional Spheres 

```python
NoDim=2       
NoPoints=1000000  
```

```python
Points=np.random.rand(NoPoints,NoDim)     
Points
```

```python
DistanceFromOrigin=np.zeros((NoPoints,1))
OutsidePoints = 0
```

```python
%%time
for U in range(NoPoints):
    for T in range(NoDim):
        DistanceFromOrigin[U]+=Points[U,T]**2
    DistanceFromOrigin[U]=np.sqrt(DistanceFromOrigin[U])
    if DistanceFromOrigin[U]>1:
        OutsidePoints+=1
```

```python
print("Fraction of Points Outside Is ",OutsidePoints/NoPoints)
```

```python
#Now i Will Do The Proble Using Numpy Functions
```

```python
%%time
SquarePoints = Points*Points
DistanceFromOrigin=np.sqrt(np.sum(SquarePoints,axis=1))   
OutsidePoints= np.sum(DistanceFromOrigin>1)
```

```python
print("Fraction of Points Outside Is ",OutsidePoints/NoPoints)
```
